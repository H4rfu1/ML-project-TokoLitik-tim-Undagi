```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.128 Safari/537.36"
chrome_options.add_argument("user-agent="+ user_agent)

def StartDriver(uri):
  wd = webdriver.Chrome('chromedriver',options=chrome_options)
  wd.implicitly_wait(30) 
  wd.get(uri)
  return wd
```

```python
url = "https://www.tokopedia.com/nourishindonesia/roasted-cashew-pieces-kacang-mede-belah-2-panggang-100-gr?whid=0"
delay = 60
target = {
    'pagination': " //*[@id='pdp_comp-review']/div[15]/div/div",
    'next_page': "//*[@id='pdp_comp-review']/div[15]/div/div/button",
    'review_block': "//p[contains(@class, 'css-1np3d84-unf-heading e1qvo2ff8')]",
    'rating_block': "//*[@id='pdp_comp-review']/div[5]/div/div[2]/div[1]",
}
```

```python
def getHeight(wd):
  max_height = wd.execute_script("return document.body.scrollHeight;") 
  # print('Total height now', max_height)
  return max_height

def scrollTo(wd, scroll=1000): # IOLazyloading
  wd.execute_script("window.scrollTo(0, "+str(scroll)+");")
  # print("current Y level",driver.execute_script("return window.scrollY;"))
  return True

def getReview(wd):
  attempts = 0;
  limit = 2;
  while (attempts < limit):
    review_block  = wd.find_elements_by_xpath(target['review_block'])
    if review_block is not None:
      break
    else:
      attempts +=1
      time.sleep(2)
  if attempts >= limit:
    print('notfound')
  for x in range (len(review_block)):
    if review_block[x].text != '':
      response_review.append(review_block[x].text)
      print(review_block[x].text)

def getPagination(wd):
  pagination  = wd.find_elements_by_xpath(target['pagination'])
  page = pagination[-1].text.split('\n')
  print(page)
  return int(page[-1])

```

```python
def changePage(wd, limit=0, current=0):
  if(current == limit):
    return False
  btn = wd.find_elements_by_xpath(target['next_page'])
  time.sleep(0.2) 
  if isinstance(btn, list):
    try:
      btn[-1].click()
    except IndexError as e:
      print(e)
    scrollTo(wd)
    getReview(wd)
  return True
      
```

```python
def crawl(uri):
  driver = StartDriver(uri)
  getHeight(driver)
  scrollTo(driver)
  page_amount = getPagination(driver)
  for i in range(page_amount):
    try:
      changePage(driver,limit=page_amount, current=i)
    except exceptions.StaleElementReferenceException as e:
      print(e)
  driver.quit()
```

# Crawl with Review

```python
response_review = {'5':[],'4':[],'3':[],'2':[],'1':[]}
```

```python
def extractRate(raw_html):
  pattern = r'icnGivenRatingFilter[0-9]+-[1-5]'
  rate = re.search(pattern, raw_html)
  return rate.group()[-1] 

def getReviewRating(wd):
  attempts = 0;
  limit = 2;
  while (attempts < limit):
    x_review = "//div[@class='css-17gqcdy']"
    review_block  = wd.find_elements_by_xpath(x_review)
    if review_block is not None:
      break
    else:
      attempts +=1
      time.sleep(2)
  if attempts >= limit:
    print('notfound')
  for x in range (len(review_block)):
    if review_block[x].text != '':
      rating = review_block[x].find_element_by_class_name('eaj5cj72')
      rating = extractRate(rating.get_attribute("outerHTML"))
      message = review_block[x].find_element_by_class_name('css-1np3d84-unf-heading')
      print(rating)
      print(message.text)
      response_review[rating].append(message.text)

```
